# Web scraping from flipkart about new apple iphones

### Importing the libraries

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
#looping through all the content in a particular page and extracting its information such as model name,price,camera quality,ratings by customers,processor
for i in range(1,11):
    url='https://www.flipkart.com/search?q=iphone&otracker=AS_Query_HistoryAutoSuggest_2_0&otracker2=00AS_Query_HistoryAutoSuggest_2_0&marketplace=FLIPKART&as-show=on&as=off&as-pos=2&as-type=HISTORY&page={}'.format(i)
    r=requests.get(url)
    soup=BeautifulSoup(r.content)
    a=soup.find_all('div',class_='_1UoZlX')
    info_names=[]
    info_price=[]
    info_ratings=[]
    camera_info=[]
    Processor_info=[]
    for i in a:
        z=i.find('div',class_='_3wU53n')
        y=i.find('div',class_='_1vC4OE _2rQ-NK')
        x=i.find('span',class_='_2_KrJI')
        w=i.find_all('li',class_='tVe95H')
        info_names.append(z.text)
        info_price.append(y.text)
        camera_info.append(w[2].text)
        Processor_info.append(w[3].text)
        
        if x is not None:
            info_ratings.append(x.text)
        else:
            info_ratings.append(int(0))
        #f.write(info_names  + info_price + info_ratings + camera_info + "\n" )
#f.close()
```
